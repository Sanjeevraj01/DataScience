**Misising value imputation**

```python
import seaborn as sns
column = DC_clean.columns
color = ['#000056','#fff000']
sns.heatmap(DC_clean[column].isnull(),
            cmap = sns.color_palette(color))
```

```python
DC_clean1.isnull().sum()
```

```python
mean = DC_clean1['Ozone'].mean()
print(mean)
```

```python
DC_clean1['Ozone'] = DC_clean1['Ozone'].fillna(mean)
```

```python
meansolar = DC_clean1['Solar.R'].mean()
```

```python
DC_clean1['Solar.R'] = DC_clean1['Solar.R'] .fillna(mean)
```

```python
monthmode = DC_clean1['Month'].mode().iloc[0]
```

```python
DC_clean1['Month'] = DC_clean1['Month'] .fillna(monthmode)
```

```python
weathermode = DC_clean1['Weather'].mode().iloc[0]
DC_clean1['Weather'] = DC_clean1['Weather'] .fillna(weathermode)
```

```python
glaxo_df= pd.read_csv('glaxo_df')
```

```python
glaxo_df
```

```python
glaxo_df_ci = stats.norm.interval(0.95,loc = glaxo_df.Close.mean(),
scale = glaxo_df.Close.std())
```

```python
np.round(glaxo_df_ci,4)
```
